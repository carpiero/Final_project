```python
list_country=final_table['Country'].unique().tolist()
final_table['Total Active']= 0
for x in list_country:
    final_table.loc[final_table['Country']==x,'Total Active']=\
     final_table.loc[final_table['Country'].str.contains(x),'Quantity'].sum()
 
final_table
```

```python
import numpy as np

final_table['Percentage']= np.where(final_table['Quantity']==0,  12345.12345,
                                    round(final_table['Quantity']/final_table['Total Active']*100))

final_table['Percentage']=np.where(final_table['Percentage']==12345.12345,
                                  'Does not apply',final_table['Percentage'].apply(lambda x: f'{int(x)} %'))

final_table['Percentage']=final_table['Percentage'].str.replace(r'0 %','less than 1 %')
                                   
```
